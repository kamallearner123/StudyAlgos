```python
class Node:
    root = None
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
    @classmethod
    def insert(cls, val):
        node = Node(val)
        if cls.root == None:
            cls.root = node
            return
        
        tmp = cls.root
        while True:
            if val < tmp.val:
                if tmp.left == None:
                    tmp.left = node
                    print(f"Val = {node.val}, to left of {tmp.val}")
                    break
                else:
                    tmp = tmp.left
            else:
                if tmp.right == None:
                    tmp.right = node
                    print(f"Val = {node.val}, to right of {tmp.val}")
                    break
                else:
                    tmp = tmp.right
                    
    @classmethod
    def inorder(cls, node=None):
        
        if node == None:
            node = cls.root
        if node == None:
            return
        
        if node.left != None:
            cls.inorder(node.left)
            
        print("Val == {}".format(node.val))
        
        if node.right != None:
            cls.inorder(node.right)
        
        return
    
    @classmethod
    def preorder(cls, node=None):
        
        if node == None:
            node = cls.root
        if node == None:
            return
        
        print("Val == {}".format(node.val))

        if node.left != None:
            cls.inorder(node.left)
        
        if node.right != None:
            cls.inorder(node.right)
        
        return
    
    @classmethod
    def postorder(cls, node=None):
        
        if node == None:
            node = cls.root
        if node == None:
            return
        
        print("Val == {}".format(node.val))

        if node.left != None:
            cls.inorder(node.left)
        
        if node.right != None:
            cls.inorder(node.right)
        
        return

    @classmethod
    def level_order(cls, node=None):
        print("In order")
        if node == None:
            node = cls.root
        if node == None:
            print("node is None")
            return
        
        Q = [node]
        
        while len(Q)>=1:
            node = Q.pop(0)
            print("Val = {}", node.val)
            if node.left != None:
                Q.append(node.left)
            if node.right != None:
                Q.append(node.right)
            
            
        
Node.insert(10)
Node.insert(1)
Node.insert(20)
Node.insert(30)
Node.insert(3)

Node.inorder()
Node.level_order()
```
